```python
def dict_to_hl7(msg_dict):
    hl7_message = []

    # Segmento MSH
    msh_segment = f"MSH|^~\&||||{msg_dict['Fecha']}|||ADT^A01|||2.6"
    hl7_message.append(msh_segment)

    # Segmento PID
    pid_segment = f"PID|||{msg_dict['Cedula']}|||{msg_dict['Fecha']}|{msg_dict['Sexo']}||{msg_dict['Apellidos']}^{msg_dict['Nombre']}"
    hl7_message.append(pid_segment)


    # Segmento OBR para cada examen
    for i, examen in enumerate(msg_dict['Examenes'], start=1):
      obr_segment = f"OBR|{i}|{examen['examen']}|||{examen['resultado']}"
      hl7_message.append(obr_segment)


    return '\n'.join(hl7_message)
```

```python
mongo_dict = {
    "Fecha": "20240129133654",
    "Cedula": "1234567890",
    "Edad": "43",
    "Nombre": "Juan Andrés",
    "Apellidos": "Sepúlveda Ramos",
    "Sexo": "M",
    "Examenes": [
        {"examen": "F", "resultado": "9.8"},
        {"examen": "A1c", "resultado": "5.9"},
        {"examen": "S-Window", "resultado": "26.2"}
    ]
}

hl7_message = dict_to_hl7(mongo_dict)
print(hl7_message)
```

```python
def guardar_hl7_en_archivo(hl7_message, nombre_archivo):
    ruta = "ruta a la carpeta"
    ruta_completa = f"{ruta}/{nombre_archivo}.txt"

    with open(ruta_completa, 'w') as file:
        file.write(hl7_message)
```
